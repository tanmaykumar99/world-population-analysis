```python
## Largest range of countries by pop:
df.sort_values(by='2022 Population',ascending=False).head()
```

```python
df.sort_values(by='2022 Population').head()
```

```python
df.corr(numeric_only = True)
```

```python
sns.heatmap(df.corr(numeric_only=True),annot=True)
plt.rcParams['figure.figsize']=(10,2) ##Figure size syntax for heatmaps
plt.show()
```

```python
##Grouping together columns to get a better understanding
pd.set_option('display.float_format',lambda x: '%.2f' % x)
df2=df.groupby('Continent')[['1970 Population','1980 Population','1990 Population','2000 Population','2010 Population','2015 Population','2020 Population','2022 Population']].mean(numeric_only=True).sort_values('2022 Population',ascending=False) ##Order by desc. order of Density
```

```python
## Need to understand 'Oceania' container
df[df['Continent'].str.contains('Oceania')].head()
```

```python
df2.plot(figsize=(15, 8))
```

```python
df3=df2.transpose()
df3.drop(['Rank','Area (km²)','Density (per km²)','Growth Rate','World Population Percentage'],inplace=True)
```

```python
df3.plot()
```

```python
## More reusbale plot
df4=df2.transpose()
df4.plot(figsize=(20,10))
```
